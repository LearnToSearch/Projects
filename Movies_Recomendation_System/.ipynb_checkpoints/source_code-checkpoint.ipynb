```python
# Importing the dependencies
```

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Data Collection and Pre-Processing
```

```python
# loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv('movies.csv')
Indian_movies_data = pd.read_csv('indian movies.csv')
```

```python
# printing the first 5 rows of the dataframe
movies_data.head(1)
```

```python
Indian_movies_data.head(1)
```

```python
# number of rows and columns in the data frame

movies_data.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null valuess with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# Cosine Similarity
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
# Getting the movie name from the user
```

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
# print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the  user 

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
# print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
# print(sorted_similar_movies)
```

```python
# Print the name of similar movies along with their genres
print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    # Fetch the title and genres from the movies_data dataframe
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    genres_from_index = movies_data[movies_data.index == index]['genres'].values[0]
    
    if i < 31:  # Limit to top 30 suggestions
        print(f"{i}. {title_from_index} - Genres: {genres_from_index}")
        i += 1
```

```python
# Movie Recommendation Sytem
```

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    # Fetch the title and genres from the movies_data dataframe
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    genres_from_index = movies_data[movies_data.index == index]['genres'].values[0]
    
    if i < 31:  # Limit to top 30 suggestions
        print(f"{i}. {title_from_index} - Genres: {genres_from_index}")
        i += 1
```
